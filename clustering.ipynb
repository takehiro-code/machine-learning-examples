# Clustering

Clustering is **unsupervised machine learning** algorithm where we don't the train the model but the model (or algorithm) by itself predicts the number without any training.

There are many clustering algorithm:

- K-Means, AgglomerativeClustering, AffinityPropagation, AffinityPropagation, DBSCAN, Hierachical Clustering, Gaussian mixtures

We demonstrated the examples of K-means and DBSCAN

See: https://scikit-learn.org/stable/modules/clustering.html

```python
# load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import normal
from sklearn.pipeline import make_pipeline
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

## K-Means Clustering

`n_clusters`: The hyperparameter to determine number of clusters.

```python
from sklearn.cluster import KMeans
# Step 1: preparing data
n = 100 # sample size
X, y = make_blobs(n_samples=[100, 100, 100, 100, 100], n_features=2, centers=None, cluster_std=0.8, random_state=0, center_box=(-10.0, 10.0))
# plt.plot(X[:,0], X[:,1], 'b.')

# step 2: choose the model
model = make_pipeline(
    KMeans(n_clusters=5)
)

## step 3: prediction
y_pred = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=5, cmap='viridis')
plt.xlabel("X[:, 0]")
plt.ylabel("X[:, 1]")
plt.show()
```

## DBSCAN

Density-Based Spatial Clustering Algorithm

`eps`: The maximum distance between two samples to be considered as a neighbor.

```python
from sklearn.cluster import DBSCAN

# Step 1: Preparing data
n = 100 # sample size
X, y = make_blobs(n_samples=[100, 100, 100, 100, 100], n_features=2, centers=None, cluster_std=1.0, random_state=0, center_box=(-10.0, 10.0))
# plt.plot(X[:,0], X[:,1], 'b.')

# step 2: choose the model
model = make_pipeline(
    DBSCAN(eps=0.5)
)

## step 3: prediction
y_pred = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=5, cmap='viridis')
plt.xlabel("X[:, 0]")
plt.ylabel("X[:, 1]")
plt.show()
```
